```python
# Function for Bonus Task
def calculate_premium(age, policy_type, claims):
    # Base premium dictionary
    premiums = {
        "health": 8000,
        "vehicle": 12000,
        "life": 10000
    }

    # Validate policy type
    if policy_type not in premiums:
        return None

    base = premiums[policy_type]
    premium = base

    # Age adjustment
    if age < 25:
        premium += base * 0.20
    elif age > 45:
        premium += base * 0.15

    # Claims adjustment
    if claims > 2:
        premium += base * 0.25
    elif claims == 0:
        premium -= premium * 0.10   # discount applied on updated premium

    return int(premium)


# Main program
name = input("Enter Customer Name: ")
age = int(input("Enter Age: "))
policy_type = input("Enter Policy Type (health/vehicle/life): ").lower()
claims = int(input("Enter Number of Claims: "))

final_premium = calculate_premium(age, policy_type, claims)

if final_premium is None:
    print("❌ Invalid policy type entered!")
else:
    base_prices = {
        "health": 8000,
        "vehicle": 12000,
        "life": 10000
    }

    print("\n--- Insurance Details ---")
    print("Customer Name:", name)
    print("Policy Type:", policy_type)
    print("Base Premium: ₹", base_prices[policy_type])
    print("Final Premium: ₹", final_premium)
```
