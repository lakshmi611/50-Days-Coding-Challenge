1. STRING Challenge
-->Longest Substring Without Repeating Characters

```python
def longest_substring(s):
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1

        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length


# Input
print(longest_substring("abcabcbb"))
```

2. LIST Challenge
--> Remove Duplicates While Preserving Order

```python
def remove_duplicates(lst):
    seen = set()
    result = []

    for num in lst:
        if num not in seen:
            seen.add(num)
            result.append(num)

    return result


# Input
print(remove_duplicates([1, 2, 2, 3, 4, 3, 5]))
```

3. TUPLE Challenge
-->Find Maximum and Minimum

```python
def find_min_max(tpl):
    return min(tpl), max(tpl)


# Input
print(find_min_max((10, 5, 20, 8)))
```

4. DICTIONARY Challenge
-->Character Frequency Counter

```python
def char_frequency(s):
    freq = {}

    for char in s:
        freq[char] = freq.get(char, 0) + 1

    return freq


# Input
print(char_frequency("programming"))
```

5. INTERMEDIATE Challenge
--> Group Words by Length

```python
def group_by_length(words):
    result = {}

    for word in words:
        length = len(word)
        if length not in result:
            result[length] = []
        result[length].append(word)

    return result


# Input
print(group_by_length(["cat", "dog", "elephant", "bat", "tiger"]))
```
