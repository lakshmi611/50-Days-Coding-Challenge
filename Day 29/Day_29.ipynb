```python
students = {
    "Asha": {"Math": 78, "Science": 85, "English": 67},
    "Ravi": {"Math": 35, "Science": 40, "English": 50},
    "John": {"Math": 90, "Science": 92, "English": 88}
}
```

1.	Calculate total & average for each student

```python
for name, marks in students.items():
    total = sum(marks.values())
    average = total / len(marks)

    print(f"{name} → Total: {total}, Average: {round(average, 2)}")
```

2.	Identify top scorer

```python
top_scorer = ""
max_total = 0

for name, marks in students.items():
    total = sum(marks.values())

    if total > max_total:
        max_total = total
        top_scorer = name

print("Top Scorer:", top_scorer)
```

3.	Find students who failed in any subject

```python
failed_students = []

for name, marks in students.items():
    if any(mark < 40 for mark in marks.values()):
        failed_students.append(name)

print("Failed Students:", failed_students)
```
