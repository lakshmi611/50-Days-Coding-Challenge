Challenge 1: Flatten a Nested List

Concept: List Comprehension

```python
nested = [[1, 2], [3, 4], [5]]

flattened = [num for sublist in nested for num in sublist]

print(flattened)
```

Challenge 2: Check Anagram

Concept: sorted()

```python
s1 = "listen"
s2 = "silent"

result = sorted(s1) == sorted(s2)

print(result)
```

Challenge 3: Find Missing Number

Concept: set()

```python
nums = [1, 2, 4, 5]

missing = list(set(range(1, 6)) - set(nums))

print(missing[0])
```

Challenge 4: Group Words by First Letter

Concept: dict, setdefault()

```python
words = ["apple", "ant", "banana", "bat"]

grouped = {}

for word in words:
    grouped.setdefault(word[0], []).append(word)

print(grouped)
```

Challenge 5: Remove Empty Values

Concept: Dictionary Comprehension

```python
data = {"a": 1, "b": None, "c": 3, "d": None}

cleaned = {k: v for k, v in data.items() if v is not None}

print(cleaned)
```
